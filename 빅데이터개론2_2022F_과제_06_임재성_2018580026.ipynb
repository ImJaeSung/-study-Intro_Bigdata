##  다음 내용은 앞의 과제에 나왔던 단순임의 추출에서의 표본 평균의 분포와 반복되는 내용입니다.  뒤에 문제를 풀 때 필요하니 파이썬 코드를 그대로 실행하세요

전체 8명인 학생들의 모임이 있다. 모임에는 여자 학생과 남자 학생이 각각 4명씩 있다. 우리가 알고 싶은 모수는 전체 8명의 키에 대한 평균(mean)이다. 

 다음은 8명의 학생들의 성별(`sex`)과 키(`height`)를 포함하는 데이터프레임 `pop`이다.  


````python
pop = pd.DataFrame({'sex': ["F","F","F","F","M","M","M","M"], 'height':[156, 153, 164, 172, 178, 187, 169, 191]})
pop
````

이제 8명 학생에서 4명을 임의로 추출할 때 모든 가능한 표본을 가진 리스트 `all_possible_samples`를 만드는 파이썬 코드를 다음과 같다.

````python
n=4

pop1 = np.array(pop.height)

all_possible_samples = []
for sample in combinations(pop1, n):
    all_possible_samples.append(sample)

all_possible_samples
````

모든 가능한 표본의 개수를 `all_possible_num` 저장한다.

````python
all_possible_num = len(all_possible_samples)
all_possible_num
````

앞에서 구한 모든 가능한 표본들을 가진 리스트 `all_possible_samples` 에서 각 표본에 대한 평균을 계산하여  데이터프레임 `sample_mean_df` 저장하는 파이썬 코드를 는 다음과 같다.

````python
sample_mean_df = pd.DataFrame({'sample_mean':np.zeros(all_possible_num)})

for i in np.arange(all_possible_num):
  sample_mean_df.loc[i, ['sample_mean']] = np.mean(all_possible_samples[i])

sample_mean_df
````

위에서 만든 데이터프레임 `sample_mean_df` 에 저장된 표본 평균들의 요약통계량은 다음과 같다.

````python
sample_mean_df.describe()
````

## 1.  층화 추출에서 표본 평균의 분포 

앞에서 8명의 학생들이 있는 `pop` 모집단을 4명의 남자 그룹과 4명의 여자 그룹으로 나누어 생각하자.  

이제 남자 그룹에서 2명, 여자그룹에서 2명을 임의로 추출한다고 하자. 

이렇게 모집단을 여러 개의 그룹 또는 층(strata)으로 나누어서 추출하는 방법을 **층화추출(stratified sampling)** 이라고 한다.    

먼저 어떻게 데이터프레임 `pop` 에서 남자와 여자 각각 2명을 임의로 추출할 수 잇을까?

다음과 같이 우리가 배운 `groupby` 메소드를 사용하면 남여 각각 2명씩 층화추출 할 수 있다. 파이썬 코드를 반복해서 실행해 보면 남여 각각 2명씩 임의로 추출되는 것을 알 수 있다.

````python
sam_st_df =  pop.groupby(by="sex").sample(n=2, replace=False)
sam_st_df
````

**문제 1**

8명의 키 자료가 있는 데이터프레임 `pop`에서 **남자의 자료만 선택하여** 데이터프레임 `pop_M`에 저장하고, 2명을 임의로 추출할 때 모든 가능한 표본을 리스트 `all_possile_samples_M` 에 저장하는 파이썬 코드를 작성하시오.



````python
pop_M = pop[pop["sex"] == "M"]
display(pop_M)

n = 2
pop_M1 = np.array(pop_M.height)

all_possile_samples_M = []
for sample in combinations(pop_M1, n):
    all_possile_samples_M.append(sample)
    
display(all_possile_samples_M)
````

**문제 2**

8명의 키 자료가 있는 데이터프레임 `pop`에서 **여자의 자료만 선택하여** 데이터프레임 `pop_F`에 저장하고, 2명을 임의로 추출할 때 모든 가능한 표본을 리스트 `all_possile_samples_F` 에 저장하는 파이썬 코드를 작성하시오.

````python
pop_F = pop[pop["sex"] == "F"]
display(pop_F)

n = 2
pop_F1 = np.array(pop_F.height)

all_possile_samples_F = []
for sample in combinations(pop_F1, n):
    all_possile_samples_F.append(sample)
    
display(all_possile_samples_F)
````

**문제 3**

위의 문제 1과 2에서 남자를 추출할 때 모든 가능한 표본들과 여자를 추출할 때 모든 가능한 표본들을 이용하여 전체 모집단에서 남여 각각 2명을 임의로 뽑는 층화 추출에서 모든 가능한 표본들을 
리스트 `all_possible_samples_st`에 저장하는 파이썬 코드를 작성히시오.


Hint: 

-  두 개의 객체 `a` 와 `b` 를 연결하여 하나의 객체로 만들어 주는 함수는 `np.concatenate()`이며 다음과 같은 파이썬 코드를 사용할 수 있다.

```
np.concatenate((a,b), axis=0)
```
-  2개의 반복문을 중첩해서 사용할 수 있다.

````python
all_possible_samples_st = []

for sample_M in all_possile_samples_M:
  for sample_F in all_possile_samples_F:
    all_possible_samples_st.append(tuple(np.concatenate((sample_M, sample_F), axis = 0)))

print(all_possible_samples_st)
````

**문제 4**

위의 문제 3에서 구한 모든 가능한 표본들을 가진 리스트 `all_possible_samples_st` 에서 각 표본의 평균을 계산하여  데이터프레임 `sample_mean_st_df` 저장하는 파이썬 코드를 작성하시오.  

Hint: `numpy` 에서 평균을 구하는 함수는 `np.mean()` 이다.

주의: 데이터프레임 `sample_mean_st_df`는 1개의 열을 가지고 있으며 열이름은 `sample_mean`이다.

````python
sample_mean = []

for i in all_possible_samples_st:
  sample_mean.append(np.mean(i))

sample_mean_st_df = pd.DataFrame({'sample_mean' : sample_mean}) 
sample_mean_st_df
````

전국의 체력인증센터에서 관리하고 있는 국민체력측정데이터의 항목별 측정 자료에서 18세이상의 성인 여자의 몸무게 자료를 다음과 같이 `numpy` 행렬 형식을 가진  `woman_weight`에 저장하였다. 

````python
url = "https://ilovedata.github.io/teaching/bigdata2/data/physical_test_2018_data.csv"
physical_data = pd.read_csv(url, sep=',', encoding = 'utf-8-sig')
````

````python
woman_weight = np.array(physical_data.loc[physical_data.TEST_AGE >= 18, "ITEM_F002"])
````

````python
n =len(woman_weight)
n
````

**문제 1**

이제 우리나라 18세 이상 전체 여성들에 대한 몸무게의 평균이 관심있는 모수 $\theta$ 라고 하자.

위에서 만든 `woman_weight` 자료가 모집단에서 단순임의추출된 표본이라고 가정하고 모수에 대한 95% 신뢰구간을 붓스트랩을 이용한 백분위수 방법으로 구하시오.

- 신뢰구간의 하한값은 `lower_bound`, 상한값은 `upper_bound`에 저장하시오.
- 붓스트랩의 모의실험 반복수는 B=500 을 사용하시오.


````python
B = 500
woman_weight_estimate = pd.DataFrame({'resample_estimate' : np.zeros(B)})

for i in np.arange(B):
  resample = np.random.choice(woman_weight, n, replace = True)
  woman_weight_estimate.loc[i, 'resample_estimate'] = np.mean(resample)

lower_bound = float(woman_weight_estimate.quantile(0.025))
upper_bound = float(woman_weight_estimate.quantile(0.975))

(lower_bound, upper_bound)
````
